```python
import numpy as np;
import pandas as pd;
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.number.unique()
```

```python
df['number_numeric'] = pd.to_numeric(df.number, errors='coerce', downcast='integer')
df['number_category'] = np.where(df['number_numeric'].isnull(), df.number,np.nan)
```

```python
df.Ticket.nunique()
```

```python
df.Cabin.unique()
```

```python
df['cabin_numeric'] = df.Cabin.str.extract('(\d+)', expand=False)
```

```python
df['cabin_category'] = df.Cabin.str[0]
```

```python
df.head()
```

```python
df.cabin_category.unique()
```

```python
df['cabin_category'].value_counts().plot(kind='bar')
```

```python
df.Ticket.unique()
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
