```python

import time
import pandas as pd
import requests

API_KEY = "8aeea0eca819e1deb50ab85647a6655b"
all_results = []

for i in range(1, 101):
    url = f"https://api.themoviedb.org/3/discover/movie?api_key={API_KEY}&include_adult=true&include_video=false&language=en-US&page={i}&sort_by=popularity.desc"
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()  # raises HTTPError if bad response
        all_results.extend(response.json().get('results', []))
    except requests.exceptions.RequestException as e:
        print(f"Error on page {i}: {e}")
        break
    time.sleep(0.3)  # wait 300ms per request (safe)

df = pd.DataFrame(all_results)
print(df.shape)


df.head(2)
```

```python
df.to_csv('movies.csv',index=False)
```
