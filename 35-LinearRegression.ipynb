```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("placement_LR.csv")
print(df.shape)
df.head()
```

```python
df.isnull().mean()
df.isnull().sum()
```

```python
plt.scatter(df.cgpa , df.package , color="red")
plt.xlabel("cgpa")
plt.ylabel("package")
plt.title("cgpa vs package")
plt.show()
```

```python
df.head()
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,1:2]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1)) , y_test.iloc[0]
```

```python
X_train
```

```python
plt.scatter(df.cgpa , df.package ,color="red")
plt.plot(X_train,lr.predict(X_train),color='blue')

plt.xlabel("cgpa")
plt.ylabel("package")
plt.title("cgpa vs package")
plt.show()
```

```python
m=lr.coef_
```

```python
b=lr.intercept_
```
