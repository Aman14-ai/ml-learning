```python
sns.barplot(data=titanic , x="Embarked" , y="Age",  hue="Sex")
```

```python
sns.boxplot(data=titanic , x="Sex" , y="Age", hue="Survived")
```

```python
sns.distplot(titanic[titanic["Survived"] == 0]["Age"], hist=False);
sns.distplot(titanic[titanic["Survived"] == 1]["Age"], 
             hist=False);
```

```python
sns.distplot(titanic[titanic["Survived"] == 0]["Pclass"], hist=False);
sns.distplot(titanic[titanic["Survived"] == 1]["Pclass"],  hist=False);
```
