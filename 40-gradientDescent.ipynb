```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)
```

```python
plt.scatter(X,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(X,y)
```

```python
lr_model.coef_
```

```python
lr_model.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,lr_model.predict(X) , color="red")
plt.show()
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = 0

y_pred = ((78.35 * X) + 0).reshape(4)

y_pred
```

```python

plt.scatter(X,y)
plt.plot(X,lr_model.predict(X) , color="red",label="OLS")
plt.plot(X,y_pred , color="green",label="GD")
plt.legend()
plt.show()
```

```python
X.ravel(),y
```

```python
m = 78.35
b = 0

loss_slope = -2 * np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
lr = 0.1
step_size = loss_slope * lr
step_size
```

```python
# calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((m * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X,lr_model.predict(X) , color="red",label="OLS")
plt.plot(X,y_pred , color="green",label="GD")
plt.plot(X,y_pred1 , color="yellow",label="GD")
plt.legend()
plt.show()
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((m * X) + b).reshape(4)
y_pred2
```
