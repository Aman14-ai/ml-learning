```python
from sklearn.datasets import make_regression
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
X , y = make_regression(n_samples=100 , n_features=1 , n_informative=1 , noise=20, random_state=13)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(X , y)
```

```python
lr = LinearRegression()
lr.fit(X , y)
y_pred = lr.predict(X)
print(lr.coef_ , lr.intercept_)
r2_score(y , y_pred)
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred,color='red')
```

```python
rdg = Ridge(alpha=10)
rdg.fit(X,y)
y_pred_rdg = rdg.predict(X)
print(rdg.coef_ , rdg.intercept_)
r2_score(y , y_pred_rdg)
```

```python
rdg2 = Ridge(alpha=100)
rdg2.fit(X,y)
y_pred_rdg2 = rdg2.predict(X)
print(rdg2.coef_ , rdg2.intercept_)
r2_score(y , y_pred_rdg2)
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred , color="red" , label="lr")
plt.plot(X,y_pred_rdg , color="blue" , label="alpha = 10")
plt.plot(X,y_pred_rdg2,color='green' , label="alpha = 100")
plt.legend()
plt.show()
```
