```python
import numpy as np;
import pandas as pd;
from sklearn.model_selection import train_test_split

from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
print("radha")
```

```python
df = pd.read_csv('train.csv')[['Age','Pclass','Fare','Survived']]
df.head()
```

```python
df.isnull().mean() * 100
```

```python
X = df.drop('Survived',axis=1)
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.head()
```

```python
knn = KNNImputer(n_neighbors=5,weights='distance')
X_train_imputed = knn.fit_transform(X_train)
X_test_imputed = knn.transform(X_test)
```

```python
lr = LogisticRegression()
lr.fit(X_train_imputed,y_train)
y_pred = lr.predict(X_test_imputed)
round(accuracy_score(y_test,y_pred)*100,2)
```

**see by mean**

```python
si = SimpleImputer(strategy='mean')
X_train_imputed_si = si.fit_transform(X_train)
X_test_imputed_si = si.transform(X_test)
```

```python
lr = LogisticRegression()
lr.fit(X_train_imputed_si,y_train)
y_pred_si = lr.predict(X_test_imputed_si)
round(accuracy_score(y_test,y_pred_si)*100,2)
```
