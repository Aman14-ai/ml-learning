```python
import numpy as np;
import pandas as pd;
```

<p>we have to do onehotencoding on gender and city cols , simple imputer on fever cols , ordinal encoding on cough col</p>

## **data analysis**

```python
df = pd.read_csv('./covid_toy.csv');
df
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer;
from sklearn.preprocessing import OneHotEncoder;
from sklearn.preprocessing import OrdinalEncoder;
from sklearn.model_selection import train_test_split;
```

```python
X_train , X_test , y_train , y_test = train_test_split(df.drop(columns=['has_covid']) , df['has_covid'] , test_size=0.2, random_state=0)
```

## **SimpleImputer()**

```python
si = SimpleImputer()
X_train_fever = si.fit_transform(X_train[['fever']]);
X_test_fever = si.transform(X_test[['fever']])
```

## **OrdinalEncoder()**

```python
oe = OrdinalEncoder(categories=[['Mild' , 'Strong']])
X_train_cough = oe.fit_transform(X_train[['cough']]);
X_test_cough = oe.transform(X_test[['cough']])
```

## **OneHotEncoder()**

```python
X_train.columns
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False)
X_train_gender_city = ohe.fit_transform(X_train[['gender' , 'city']])
X_test_gender_city = ohe.transform(X_test[['gender' , 'city']]);
```

# **Mentos Life using column transformer**

```python
from sklearn.compose import ColumnTransformer;
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1' , SimpleImputer() , ['fever']),
    ('tnf2' , OrdinalEncoder(categories=[['Mild', 'Strong']]) , ['cough']),
    ('tnf3' , OneHotEncoder(drop='first', sparse_output=False) , ['gender' , 'city'])
]  , remainder = 'passthrough');
```

```python
transformer.fit_transform(X_train).shape
```

```python
transformer.transform(X_test).shape
```
